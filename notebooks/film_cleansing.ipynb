```python
import pandas as pd
import numpy as np


    
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
pd.set_option('display.max_rows', None)  # Mostrar todas las filas
pd.set_option('display.max_colwidth', None) # Mostrar el contenido completo de las celdas
```

```python
df2 = pd.read_csv("C:/Users/luiso/IRONHACK/w3-project_SQL/data/film.csv")

df2.head()
```

```python
dim_original = df2.shape
film = df2.copy()
film.shape, dim_original
```

```python
film.info()

film.isnull().sum()
```

```python
film.duplicated().sum()
```

```python
# Eliminar la columna "oriiginal_language_id" porque no aporta información

film.drop(columns = ["original_language_id"], inplace = True)
```

```python
film.head()
```

```python
# Vamos a crear 3 columnas nuevas con los valores de special_features

film["Behind_the_Scenes"] = film["special_features"].str.contains("Behind the Scenes")
film["Commentaries"] = film["special_features"].str.contains("Commentaries")
film["Deleted_Scenes"] = film["special_features"].str.contains("Deleted Scenes")

# Especificamos que las peliculas que no tienen ninguno de estos valores, se les asigna un False a las no la tienen y True a las que si tienen

film["Behind_the_Scenes"] = film["Behind_the_Scenes"].fillna(False)
film["Commentaries"] = film["Commentaries"].fillna(False)
film["Deleted_Scenes"] = film["Deleted_Scenes"].fillna(False)

# Ahora quiero sustituir los valores True y False por "Included" y "Not_included" respectivamente a traves de una función

def special_features(x):
    if x == True:
        return "Included"
    else:
        return "Not_included"

# Aplicamos la función a las 3 columnas nuevas

film["Behind_the_Scenes"] = film["Behind_the_Scenes"].apply(special_features)
film["Commentaries"] = film["Commentaries"].apply(special_features)
film["Deleted_Scenes"] = film["Deleted_Scenes"].apply(special_features)








film.head()
```

```python
# Eliminamos la columna "special_features" porque ya no nos aporta información

film.drop(columns = ["special_features"], inplace = True)
```

```python
film.head()
```

```python
# Cambiar los nombres de las columnas a minúsculas

film.columns = film.columns.str.lower()

film.head()
```

```python
# Exportar a csv

film.to_csv("C:/Users/luiso/IRONHACK/Blockbuster_Project_2/film_clean.csv", index = False)
```
